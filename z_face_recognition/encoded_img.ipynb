```python
import base64
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Đọc ảnh thành một chuỗi bytes

```python
with open("E:/BKDN/ky6/PBL5/src/z_face_recognition/NguyenAnh_test.jpg", "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    print(encoded_string)
```

test: https://codebeautify.org/base64-to-image-converter

## Decode chuỗi base64 thành ảnh

```python
decoded_image = base64.b64decode(encoded_string)
print(decoded_image)
```

## Hiển thị ảnh bằng OpenCV

```python
img_array = np.frombuffer(decoded_image, np.uint8)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

## Hiển thị ảnh bằng Matplotlib

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
